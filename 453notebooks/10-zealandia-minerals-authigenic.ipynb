# Zealandia authigenic minerals


*https://en.wikipedia.org/wiki/Manganese_nodule*

Chemicals are dissolved from the land and transported to the sea by rivers; or hydrothermal processes dissolve the seabed in special environments (e.g. arcs, mid-ocean ridges, or other volcanic features). Hydrological cycles concentrate certain elements in the ocean (why it is salty). In parts of the ocean with low sedimentation rate and a steady supply of enriched bottom-water, saturation may cause authigenic minerals to precipitate (the definition of 'authigenic') either directly from seawater (hydrogenous nodules) or through diagenetic reactions with sediment involved (diagenetic nodules). 

The seabed of southern New Zealand is modified by the largest current system on Earth: the Antarctic Circumpolar Current (ACC). Shallow water of the Chatham rise (200-2000 m) is encrusted with phosphate, and deeper water (2000-5000 m) at the southeastern margin of the Campbell Plateau is covered with manganese crust and nodules (about 500,000 sq. km; see map below; distribution from Bostock et al. 2019).



## Materials to enable global energy transition

In previous notebooks we looked at a scenario for future energy consumption, and how we might transition to solar and wind supplies, but noted factors such as raw materials, technological barriers, and human behaviour could slow the transition.  

Availability of earth materials may limit large-scale uptake of alternatives to fossil fuels. We will need: copper (Cu) for wires; lithium (Li) and cobalt (Co) for batteries; and rare earth elements (REE) for magnets, metal alloys, and other technologies. It has been suggested that manganese (Mn) might be used to construct cheaper lower-maintenance batteries than lithium, and based on our previous analysis of lithium it seems likely that alternatives will be needed for some applications (e.g. industrial storage where weight is not important, see Chen et al., 2018 for use of Mn for this).  

We are not aware of any plan to mine manganese nodules south of New Zealand. However, in this notebook we consider the Mn, Co, and REE near the southern Campbell Plateau. Global Co and REE production is entirely from the Democratic Republic of the Congo and China, respectively, so there are already supply-chain issues. The primary ingredients of the nodules (manganese and iron) may not be economic to mine on their own, but the trace metals in the nodules may add considerable value.

#### What is the total size of the resource we have in this nodule field?  
(Disclaimer: this is a very rough calculation.)

```python
# USGS 2020 statistics 
# https://pubs.usgs.gov/periodicals/mcs2020/mcs2020-manganese.pdf
# https://pubs.usgs.gov/periodicals/mcs2020/mcs2020-cobalt.pdf
# https://pubs.usgs.gov/periodicals/mcs2020/mcs2020-rare-earths.pdf
Mn_globalReserves_tons_2019 = 810e9
Mn_globalProduction_tons_2019 = 19e6
Co_globalReserves_tons_2019 = 7e6
Co_globalProduction_tons_2019 = 140e3
REE_globalProduction_tons_2019 = 210e3
REE_globalReserves_tons_2019 = 120e6

# Weight proportion values for Campbell Plateau nodules
# Based on Davies 2018, Chang et al. 2003, Carter 1989
Mn_prop = 0.19
Fe_prop = 0.15
Co_prop = 0.002
REE_prop = 0.001

nodulesTonsPerSqM = 0.3
nodulesTonsPerSqKm = nodulesTonsPerSqM * 1e6

totalSurfaceAreaSqKm = 500000

Mn_tons_total  = Mn_prop  * totalSurfaceAreaSqKm * nodulesTonsPerSqKm
Co_tons_total  = Co_prop  * totalSurfaceAreaSqKm * nodulesTonsPerSqKm
REE_tons_total = REE_prop * totalSurfaceAreaSqKm * nodulesTonsPerSqKm

print('Mn  TOTAL RESERVES (millions of tonnes):', Mn_tons_total/1e6)
print('Co  TOTAL RESERVES (millions of tonnes):', Co_tons_total/1e6)
print('REE TOTAL RESERVES (millions of tonnes):', REE_tons_total/1e6)
print('Mn  TOTAL RESERVES (multiple of existing global reserves):', Mn_tons_total/Mn_globalReserves_tons_2019)
print('Co  TOTAL RESERVES (multiple of existing global reserves):', Co_tons_total/Co_globalReserves_tons_2019)
print('REE TOTAL RESERVES (multiple of existing global reserves):', REE_tons_total/REE_globalReserves_tons_2019)
```

```python
areaMinedPerYear = 0.5 * Co_globalProduction_tons_2019 / (Co_prop * nodulesTonsPerSqKm)

print('To add 50% to global Co production requires mining',areaMinedPerYear,'sq km/yr.')
print('After 30 yr, this amounts to',
      100 * 30 * areaMinedPerYear/totalSurfaceAreaSqKm, '% of the total area.')

print('\nAnnual market value of finished products (M USD / yr)')
print('Co  ',Co_prop  * nodulesTonsPerSqKm * areaMinedPerYear * 34000/1e6)
print('REE ',REE_prop * nodulesTonsPerSqKm * areaMinedPerYear * 8100 / REE_globalProduction_tons_2019)
print('Mn  ',Mn_prop  * nodulesTonsPerSqKm * areaMinedPerYear * 5/1e6)
```

The values are significant in comparison to the total size of New Zealand GDP (about 4% of 200B$/yr; maybe a bigger plan is possible?). Mining would target the richest areas, so the proportion of total area affected would likely be less than estimated above, which uses average values. However, this plan ignores the significant difficulty of processing the ore, which is highly problematic.


## Mining would damage fragile environments

*Image modified from FFI 2020.*

Seabed mining would destructively modify the local environment and there is no way to mitigate that. Sediment from mining would be deposited downstream (to the north) and that would also modify the seabed. The affected area represents only a small part of a very large deep-water environment (that is not rich in life, but may have unique species) and so it is unlikely to make any species extinct.    

#### What about plumes of sediment that would affect shallow water in a wider area?
The area affected is remote from land and so it represents only a very small part of the surrounding pelagic ecosystem, but there likely would be some disruption to local life. The plume of Fe-Mn sediment waste would provide nutrients and would fertilize the ocean, likely triggering significant phytoplankton blooms (Boyd et al. 2000). The net result may in fact have two signifiant benefits: (1) stimulation of the ecosystem (more fish, etc.); and (2) a large net draw-down of carbon dioxide.

## National priorities and international regulation

The United Nations Convention on the Law of the Sea (UNCLOS) defines sovereignty of the seabed, and most of this nodule field lies just within New Zealand (the western end is Australian, and some is a marine reserve). Maybe it is not economic here or is not politically acceptable in New Zealand (have we said NO to seabed mining?). However, significant areas of deep seabed elsewhere in the ocean could be mined with little or no oversight, because they are international waters. This is becoming a significant issue that needs to be resolved - companies could start mining anyway with no political oversight or regulation.

### Farming
Natural plants and animals can exist sustainably for very long periods of time. However, our farms have to apply large amounts of fertilizer every year to maintain their productivity. Is that sustainable? Is it desirable? Below shows the mass of phosphorous (P) and nitrogen (N) applied each year in New Zealand.

```python
import numpy as np
import matplotlib.pyplot as plt

dataDir = '../DATA/fertilizer/'
filename = dataDir + 'PNfertilizerUseNZ.csv'

year,tonsN,tonsP = np.genfromtxt(filename, delimiter=',', skip_header=4,
                       usecols=(0,1,2), unpack=True, dtype=float)

plt.figure(figsize=[10,5])
plt.plot(year, tonsP/1000, label='Phosphorous')
plt.plot(year, tonsN/1000, linestyle='--', label='Nitrogen')
plt.legend()
plt.xlabel('Year')
plt.ylabel('Elemental mass of fertilizer (1000 metric tons)')
plt.show()

# Chatham Rise phosphate
layerThickness_m = 0.2
density_kgPerCuM = 3000
propP = 0.09
massP_kgPerSqM = propP * layerThickness_m * density_kgPerCuM
massP_1000tonPerSqKm = massP_kgPerSqM
print('Surface area (sq km) to mine to meet annual P demand of 150,000 tons:', 150/massP_1000tonPerSqKm)
```

Nitrogen can be extracted from the air, so it is sustainable to obtain it, but it is leached into groundwater and rivers, affecting water quality for drinking and ecosystems. Phosphate is mined. Most of our phosphate comes from west Africa where there are poor outcomes for local people and the environment. If we mined it locally (Chatham Rise), we would need to mine about 3 sq km per year.
