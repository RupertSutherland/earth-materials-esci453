# Exploration, appraisal, reserves, production, remediation  

Earth materials generally go through various stages of maturity. Initially, a resource is found by **exploration** and more detailed and expensive exploration allows delineation of a potential asset. Exploration that satisfies sufficiently strict criteria to allow delineation of the amount of a producible resource is called **appraisal**. The amount of asset that could be **produced** is called a **reserve**. When production is finished, the site of production must be **remediated** back to acceptable environmental conditions, as agreed by a local regulator or government, or/and in accordance with company policy.  

## Reserve definitions  
The **proven reserve** of a natural resource is defined as the quantity estimated with reasonable certainty, from analysis of geologic and engineering data, to be recoverable from well established or known sources with existing equipment and operating conditions.  

In the oil industry, proven reserves (also referred to as P1 or P90) are classified as having a 90% or greater likelihood of being present and economically viable for extraction in current conditions. Similar rules exist for the minerals industry. Proven reserves take into account the technology used for extraction, regional regulations, and market conditions. Hence, proven reserves can take sudden leaps and drops. Extraction companies usually have to disclose proven reserves (financial assets) due to regulations, but keep as much information as they can secret to give themselves competitive advantage.  

**Technically recoverable** reserves may be known to exist with high certainty but are not economically profitable (with existing technology). They are not included in proven reserves. 

**Probable reserves** have a 50-90% probability of being present and economic. They are often excluded from reserve estimates.  

**Possible reserves** plausibly exist, but with a less than 50% probability. This usually reflects a lack of exploration and appraisal data that reduce uncertainty. The are almost never included in reserve estimates.  

**Key points**
* "Reserves" are usually proven reserves, which are a small subset of possible reserves.
* Proven reserves change due to exploration and appraisal programs AND economic, technology and regulatory conditions.
* Investment in exploration and appraisal only occurs when future production becomes threatened by availability (not before).



## UK coal production  

The industrial revolution started in the UK and was based on energy derived from burning locally-mined coal. 

```python
import numpy as np
import matplotlib.pyplot as plt

# You can change the runtime configuration of pyplot (customize styles) using rc() or rcParams()
# https://matplotlib.org/stable/tutorials/introductory/customizing.html
# e.g. force use of TrueType fonts in pdfs for ease of later editing
from matplotlib import rc
rc("pdf", fonttype=42)

# You can also use style sheets to get a consistent look if you don't like defaults
# See https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html
plt.style.use('seaborn')

filename = '../data/coal/coal-output-uk-tonnes.csv'

year,production,imports = np.genfromtxt(filename,delimiter=',',
                                        skip_header=1,usecols=(2,3,4),unpack=True)

plt.figure(figsize=[6,4])
plt.title('A',loc='left')

plt.plot(year,production/1e6,label='UK coal production',color='black',linestyle=':')
plt.plot(year,imports/1e6,label='UK coal imports',color='red',linestyle='-')

plt.xlabel('Year')
plt.ylabel('Coal (millions of tonnes/year)')
plt.legend()
plt.savefig('../figs/uk-coal.jpg',dpi=300)
plt.savefig('../figs/uk-coal.pdf')
plt.show()
```

The peak of coal production and use in the UK was more than a century ago. Environmental considerations associated with toxic smoke from coal burning, combined with the greater utility of oil, and then the discovery of oil and natural gas in the nearby North Sea caused the UK to abandon coal as its primary energy source. Climate change considerations were not influential, but were probably a reason that contributed to abandoning cheap imports of coal. There is still coal in the ground in the UK (large 'possible reserves'), but why would anybody spend money surveying it (to turn it into 'proven reserves') when it is not wanted?  
**Resource utilization and the size of reserves depends more on demand than geology**
